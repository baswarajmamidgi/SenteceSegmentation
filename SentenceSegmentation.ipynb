```python
from bs4 import BeautifulSoup
import nltk
import urllib.request
path='https://en.wikipedia.org/wiki/Elon_Musk'
response=urllib.request.urlopen(path)
soup = BeautifulSoup(response, 'html.parser')
#print(soup.prettify())  prints entire source code

print(soup.title) #prints title

```

```python

data=soup.get_text()  #entire data present in page

```

```python
from nltk.tokenize import TweetTokenizer  #for tokenizing the entire text
tknzr=TweetTokenizer()
tknzr.tokenize(data)
```
